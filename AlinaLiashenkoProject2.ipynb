```python
#imported all required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import numpy as np
import io #io-name of library
from google.colab import files as f #google colab - set of libraries, files-certain library
```

```python
uploaded = f.upload() #function to load file from local computer 
```

```python
df = pd.read_csv(io.BytesIO(uploaded['NYPD_Arrest_Data__Year_to_Date_.csv'])) #uploaded dataset 
df
```

```python
df.isnull().sum() #checked and count missing values
```

```python
df.drop(labels=['PD_CD', 'KY_CD', 'LAW_CAT_CD'], axis=1) #removed columns with missing data, since they will not be included in the analysis
```

```python
df['PD_DESC'].value_counts() #counted the number of crimes by type
```

```python
df1=df['PD_DESC'].value_counts().nlargest(10) #pointed out the top 10 type of crimes by numbers
df1
```

```python
df1=df['PD_DESC'].value_counts().nlargest(10).plot(kind='barh') #created 'barh' plot for the top 10 type of crimes by numbers
df1
```

```python
df.groupby('ARREST_BORO')['PD_DESC'].count() #used groupby function to find the number of crimes by borough 
```

```python
df.groupby(['ARREST_BORO'])['PD_DESC'].count().sort_values(ascending=False).plot(kind='bar', color=['black', 'red', 'green', 'blue', 'cyan']) #created bar plot to show the number of crimes by borough 
```

```python
df.groupby('PERP_SEX')['PD_DESC'].count() #used groupby function to find the number of crimes committed by men and women
```

```python
df.groupby(['PERP_SEX'])['PD_DESC'].count().sort_values(ascending=False).plot(kind='bar', color=['blue', 'red']) #created bar plot to show the number of crimes committed by men and women
```

```python
df.groupby('PERP_RACE')['PD_DESC'].count() #used groupby function to find the number of crimes committed by race 
```

```python
df.groupby(['PERP_RACE'])['PD_DESC'].count().sort_values(ascending=False).plot(kind='bar', colormap='viridis') #created bar plot to show the number of crimes committed by race 
```
