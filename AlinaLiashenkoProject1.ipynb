```python
missing_values = ["s", "na"] #pointed out unexpected missing values
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv', na_values = missing_values) #uploaded updated dataset with new values for missing data
df
```

```python
df.isnull().sum() #checked and count missing values
```

```python
df=df.dropna(how='all') #deleted all missing values, if all cells in the raw with missing values - totally empty 
df
```

```python
df=df.dropna(subset=['Mean Score']) #deleaded rows if values in the column 'Mean score' is missed, this value will be used for analytics 
df
```

```python
df1=df[(df["School Name"] == 'Orchard Collegiate Academy') | (df["School Name"] =='Hyde Leadership Charter School')] #|or sign, created a subset with 2 schools for analysis
df1
```

```python
df1.groupby(['Regents Exam','School Name'])['Mean Score'].median().plot.bar() #created classical bar plot
df1
```

```python
sns.barplot(x = 'Mean Score', y = 'Regents Exam', hue = 'School Name', data = df1) #created bar plot to see the difference in 'Mean score' between two selected schools
```

```python
sns.barplot(x = 'Percent Scoring 80 or Above', y = 'Regents Exam', hue = 'School Name', data = df1) #created bar plot to see the difference in 'Percent Scoring 80 or Above' between two selected schools
```

```python
sns.barplot(x = 'Percent Scoring Below 65', y = 'Regents Exam', hue = 'School Name', data = df1) #created bar plot to see the difference in 'Percent Scoring Below 65' between two selected schools
```

For analysys were selected 2 independed schools from NY in different boro - Orchard Collegiate Academy (Manhattan) and Hyde Leadership Charter School (Bronx).
**Conclusions:**
1. The analysis shows that *Mean score* at Hyde Leadership Charter School is higher for 10 out of 11 compared subjects. Also, this school provides 3 additional subjects (or do not passed such tests) as 'Algebra /Trigonometry', 'Physical settings/Chemistry', and 'Physical settings/Physics'. 
2. *Percent Scoring 80 or Above* at Hyde Leadership Charter School higher than in Orchard Collegiate Academy for 8 subjects. Orchard Collegiate Academy leads for 3 subjects: 'Geometry', 'Integrated Algebra', and 'Physical Settings/Earth Science'.
3. *Percent Scoring Below 65* at Orchard Collegiate Academy higher than in Hyde Leadership Charter School for 8 subjects.

Recording this data can be said about the higher academic achievement of Hyde Leadership Charter School students.
